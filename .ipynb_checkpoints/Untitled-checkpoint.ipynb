### A6 


```python
import math

def kiemtrasonguyento(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3,n):   
        if n % i == 0:
            return False
    return True

a, b = map(int, input().split(","))

total = 0
for x in range(a, b + 1):
    if kiemtrasonguyento(x):
        total += x

print(total)
```

### A8

```python

#str de doi n tu int thanh str 
#(n)[::-1] viet nguoc lai n
def reverse(a):
    a=int(str(a)[::-1])
    return(a)
def palindrome(b):
    return str(b)==str(b)[::-1]
n= int(input())
buoc=0
while not palindrome(n):
    n=n+reverse(n)
    buoc=buoc+1
print(buoc, n)    
        
```
